### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Exercise 1 Answer:

$ \frac{\partial u_g}{\partial x} + \frac{\partial v_g}{\partial y} = \frac{\partial}{\partial x}(\frac{-1}{f}\frac{\partial \Phi}{\partial y}) + \frac{\partial}{\partial y}(\frac{1}{f}\frac{\partial \Phi}{\partial x}) = (\frac{-1}{f})\frac{\partial^2 \Phi}{\partial x \partial y} + \frac{\partial}{\partial x}(\frac{-1}{f})\frac{\partial \Phi}{\partial y} + (\frac{1}{f})\frac{\partial^2 \Phi}{\partial x \partial y} + \frac{\partial}{\partial y}(\frac{1}{f})\frac{\partial \Phi}{\partial x} = 0 + \frac{\partial}{\partial y}(\frac{1}{f})\frac{\partial \Phi}{\partial x} = 0 $ if $\beta = df/dy =0$


#### Key points: the geostrophic wind is boring for 2 reasons: because it is not connected to vertical velocity $\omega$, and also because it would never change with time since its definition sets d/dt to zero. The implication is that ageostrophic wind, despite its relatively small magnitude, is especially *interesting*. However, the geostrophic wind is adequate to approximate the direction and speed of advection of various scalar fields (temperature, vorticity) in the atmosphere.

## Answer to exercise 3: Acceleration-related ageostrophic winds 

Consider an air parcel following the red true wind vectors along the trough pictured above. The parcel is moving counterclockwise, so its acceleration is towards the center of the area of low pressure. So, since the Coriolis force of the ageostrophic wind is to the right of the yellow ageostrophic wind vectors, this force is in the same direction as the acceleration of the parcel.  For instance, at the southernmost point of the trough where the magnitude of the ageostrophic wind is largest, the winds are accelerating towards the north, while the ageostrophic winds are pointing to the east. The Coriolis force of the ageostrophic wind is pointing to the right of the ageostrophic wind, or toward the north. Thus, the acceleration of the parcel is in the same direction as the Coriolis force of the ageostrophic wind.

### Background

The sense of the QG omega equation, if second derivatives are treated like fancy negative signs and coefficients like static stability are treated as constants, is: 

** (Upward motion) = (Vorticity advection increasing w/ height) + (Warm advection) **

Typically, low-level vorticity advection is weak, so that one can simply interpret *upper level (like 500 or 200 hPa) vorticity advection* as an indicator of the first term. 


Notice that the QG omega equation is a ***diagnostic*** equation, not a **prognostic** equation with a time derivative in it. That is, the right-hand terms do not "lead to" upward motion, they are *equal to it, right now*. 

How did this equation come from our governing Primitive Equations set, which has a prognostic form: *d/dt( {winds, and thickness-related T} )  = {PGF&Coriolis, and physical heating rate Q} )*?

The physical concept behind the QG omega equation is, somehow, to state or assume or express the condition that *the flow remains balanced at all times* -- in other words, $\partial (imbalance)/ \partial t$ = 0. That is, the QG "(Upward motion)" in the equation above is *the hypothetical vertical motion field that exactly compensates the balance-destroying tendencies* of the terms on the right-hand side

## Exercise 0 answer: 


The Q-G omega equation was derived from the horizontal momentum equations and the thermodynamic energy equation. It takes into account the breaking down of the thermal wind balance through the advection of u,v, or T. The Q-G omega equation explains that vertical velocity is just what is needed to restore the thermal wind balance.

South/southwest of the storm's center, there is counterclockwise turning (i.e. backing) of the winds with height, which corresponds to cold advection. North/northeast of the storm's center, there is clockwise turning (i.e. veering) of the winds with height, which corresponds to warm advection. The temperature advection magnitudes are largest where the surface pressure and height contours are most nearly perpendicular. The second image depicts the regions of positive (red) and negative (blue) vorticity advection. The animation of absolute vorticity shows that the regions of PVA and NVA are reasonable, as the regions of absolute vorticity are propagating northeastward with the system. The final image shows the vertical velocity, with purple referring to rising motion and green referring to sinking motion. This follows the Q-G omega equation, since there is rising motion where there is PVA and warm advection, and sinking motion where there is NVA and cold advection. Similar to the example from previous exercises, it seems that the vorticity advection term carries more weight on the vertical motion than the temperature advection term.
